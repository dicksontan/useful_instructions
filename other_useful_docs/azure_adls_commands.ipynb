```python
import paramiko

class SFTPConnector:
    def __init__(self, host_name, username, password):
        self.host_name = host_name
        self.username = username
        self.password = password
        self.transport = None
        self.sftp = None
    
    def connect(self):
        self.transport = paramiko.Transport((self.host_name, 22))
        self.transport.connect(username=self.username, password=self.password)
        self.sftp = paramiko.SFTPClient.from_transport(self.transport)

    def download_file(self, remote_file_path, local_file_path):
        self.sftp.get(remote_file_path, local_file_path)

    def upload_file(self, local_file_path, remote_file_path):
        self.sftp.put(local_file_path, remote_file_path)

    def delete_file(self, remote_file_path):
        self.sftp.remove(remote_file_path)

    def create_directory(self, remote_directory):
        self.sftp.mkdir(remote_directory)

    def delete_directory(self, remote_directory):
        self.sftp.rmdir(remote_directory)

    def make_directory(self, remote_directory,mode):
        self.sftp.mkdir(remote_directory,mode)

    def close(self):
        if self.sftp:
            self.sftp.close()
        if self.transport:
            self.transport.close()

# host_name="zpdataengnrsftp.blob.core.windows.net",
# username="zpdataengnrsftp.organon.deuser",
# password="4T+lHjtKSNL2J9eWqNshaKNW9y5y/lXy",

# Create SFTP connector object
sftp_conn = SFTPConnector(
    host_name="zpdataengnrsftp.blob.core.windows.net",
    username="zpdataengnrsftp.vm-migration-dev.deuser",
    password="4T+lHjtKSNL2J9eWqNshaKNW9y5y/lXy"
)

# Connect to SFTP server
sftp_conn.connect()

# For example, you can list files in a directory:
files = sftp_conn.sftp.listdir('./datafeed/')
print("Files in current directory:", files)

# #download files
# remote = r'./datafeed/abbott-nutriton/Zuellig_20240408.csv'
# local=r'.test10.csv'
# sftp_conn.download_file(local_file_path=local,remote_file_path=remote)

#put files
remote_directory=r'./datafeed/abbott-nutriton/a/b'
sftp_conn.make_directory(remote_directory, mode=755)
remote = r'./datafeed/abbott-nutriton/a/b/output1.csv'
local=r'C:\Users\dtands\Desktop\z_vs_code_projects\output.csv'

sftp_conn.upload_file(local_file_path=local,remote_file_path=remote)

# remote = r'./datafeed/abbott-nutriton/output1.csv'
# sftp_conn.delete_file(remote_file_path=remote)

#sftp_conn.create_directory('./datafeed/abbott-nutriton/remote_directory')

# sftp_conn.delete_directory('./datafeed/abbott-nutriton/remote_directory')

# files = sftp_conn.sftp.listdir('./datafeed/abbott-nutriton')
# print("Files in current directory:", files)

# Remember to close the connection when you're done
sftp_conn.close()
```
